## For

````python
#for
for contador in range(10):
  print(contador)
````

````python
#for + break + break
for contador in range(500):
  print(contador)
  if contador > 8 :
    break
````

````python
#for simples
lista = ["lucian", "gabriel", "jose"]
for val in lista:
  print(val)
````

````python
# for com enumerate
for i, val in enumerate(lista):
  print(i,val)
````
